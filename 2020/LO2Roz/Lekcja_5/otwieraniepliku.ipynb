```python
# tworzenie liczb losowych
from random import randint
```

```python
print(randint(1,20))
```

```python
list = []
for i in range(40):
    list.append(randint(1,40))
print(list)
```

```python
# sortowanie liczb
order = True
e = 0
while order == True:
    order = False
    for i in range(len(list) - 1):
        if list[i] > list[i+1]:
            a = list[i]
            b = list[i+1]
            list[i+1] = a
            list[i] = b
            order = True
    e+=1
print(list)
print(e)
```

## Otwieranie pliku Excel ##

```python
import pandas as pd
```

```python
df = pd.read_excel('lista plac payroll.xlsx')
```

```python
df
```

```python
df['Kwota podatek'] = df['wynagrodzenie netto']*df['podatek']
df['Kwota ubezpieczenie'] = df['wynagrodzenie netto'] * df['ubezpieczenie']
df['kwota brutto'] = df['wynagrodzenie netto']+df['Kwota podatek'] + df['Kwota ubezpieczenie']
df
```

```python
x = df['wynagrodzenie netto'].mean()
print(x)
```

```python
def ocena(s):
    if s > x:
        return 'wynagrodzenie wyzsze'
    return 'wynagrodzenie nie większe'
```

```python
df['Ocena'] = df['wynagrodzenie netto'].apply(ocena)
df
```

```python
df[df['Ocena']=='wynagrodzenie wyzsze']
```

```python
df.to_excel('zapis.xlsx', sheet_name='payroll')
```
