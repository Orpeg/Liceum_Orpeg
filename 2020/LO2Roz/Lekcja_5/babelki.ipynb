**Zadanie 33**

utworzyć zbiór liczbowy nieuprządkowany z zakresu od 1 do 20 i dokonać jego uporządkowania rosnąco metodą bąbelkową.

```python
list = []
for e in range(20):
    list.append(randint(1,20))
print(list)
```

```python
order = True
while order:
    order = False
    for e in range(len(list)-1):
        if list[e] > list[e+1]:
            a = list[e]
            b = list[e+1]
            list[e+1] = a
            list[e] = b
            order = True
print(list)
```

** Wymaga użycia biblioteki pandas **

```python
import pandas as pd
```

```python
df = pd.read_excel('lista plac payroll.xlsx')
```

```python
df
```

```python
df['Kwota podatek'] = df['wynagrodzenie netto']*df['podatek']
df
```

```python
df['Kwota ubezpieczenie'] = df['wynagrodzenie netto']*df['ubezpieczenie']
df
```

```python
df['BRUTTO'] = df['wynagrodzenie netto']+df['Kwota podatek']+df['Kwota ubezpieczenie']
df
```

```python
df['wynagrodzenie netto'].mean()
```

```python
def srednia(x):
    if x > 1835.85:
        return 'większa'
    return 'mniejsza lub równa'
```

```python
df['ocena'] = df['wynagrodzenie netto'].apply(srednia)
df
```

```python
df.loc[df['ocena'] == 'większa']
```

```python
df.to_excel('payroll modyfikacja.xlsx', sheet_name='lista imienna')
```
